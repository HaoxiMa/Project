```python
data = pd.read_table("training_set.txt", sep=",")
data.head()
test = pd.read_table("test_set.txt", sep=",")
data.head()
```

```python
def sigmoid(z):
    h = 1 / (1+np.exp(-z))
    return h
```

```python
cols = data.shape[1]
X = data.iloc[:,1:cols]
y = data.Buy
# convert to numpy arrays and initalize the parameter array theta
X = np.array(X.values)
y = np.array(y)

#test seet
cols = test.shape[1]
Xval = test.iloc[:,1:cols]
yval = test.Buy
# convert to numpy arrays and initalize the parameter array theta
Xval = np.array(Xval.values)
yval = np.array(yval)
```

```python
X.shape,y.shape,Xval.shape,yval.shape
```

```python
def cost(theta, X, y, lamda):
    first = (-y) * np.log(sigmoid(np.dot(X,theta)))#   * 表示array对应元素相乘
    second = (1 - y)*np.log(1 - sigmoid(np.dot(X,theta)))
    part1 = np.mean(first - second)
    part2 = (lamda/(2*len(X)))*(theta[1:]@theta[1:])#一维array内积
    return part1+part2
```

```python
def gradient(theta,X,y,lamda):
    grad = (X.T @ (sigmoid(np.dot(X,theta))-y))/len(X)#   @ 表示array的矩阵乘法和np.dot()是一样的
    reg = (lamda/len(X))*theta
    reg[0] = 0
    return grad + reg
```

```python

def trainLogReg(X,y,lamda):
    theta = np.zeros(X.shape[1])
    result = opt.minimize(fun=cost, x0=theta, args=(X,y,lamda),method="TNC",
                         jac=gradient) 
    return result["x"]
```

```python
def Learning_curves(X,y,Xval,yval,lamda):
    training_cost, cv_cost = [], []
    seq = range(100,len(X)+1)
    for i in seq:
        theta = trainLogReg(X[0:i,:],y[0:i],lamda)
        training_cost_i = cost(theta,X[0:i,:],y[0:i],0)#lamda=0 
        training_cost.append(training_cost_i)
        cv_cost_i = cost(theta,Xval,yval,0)
        cv_cost.append(cv_cost_i)
    fig, ax = plt.subplots(figsize=(8,8))
    ax.plot(seq,training_cost,label="training cost")
    ax.plot(seq,cv_cost,label="cv cost")
    ax.legend()
    ax.set_xlabel("Number of training examples")
    ax.set_ylabel("Error")
    ax.grid(True)
    plt.title("Learning curves")
```

```python
Learning_curves(X,y,Xval,yval,0)
```
